```python
# Fonctions pour convertir une colonne en type float plus remplire les valeurs manquantes (nan) avec la moyenne de cette colonne
def toNumericAndMean(dataFrame, column):
    dataFrame[column] = pd.to_numeric(dataFrame[column])
    values=dataFrame[column].mean()
    dataFrame[column].fillna(value=values, inplace=True)
```

```python
# Nous indique le nombre de valeur unique d'une colonne donnée
def countUnique(dataFrame, column):
    print(np.count_nonzero(dataFrame[column].unique()))

countUnique(df, 'Item_Fat_Content')
```

```python
# Modifie les strings non voulu d'une colonne donné avec les strings souhaitées
def removeAndModifyUnwantedString(dataFrame, column):
    for i in range(len(dataFrame[column])):
        if(dataFrame[column][i] == 'low fat' or dataFrame[column][i] == 'LF'):
            dataFrame[column][i] = 'Low Fat'
        elif dataFrame[column][i] == 'reg' :
            dataFrame[column][i] = 'Regular'
        elif dataFrame[column][i] == 'None': 
            dataFrame[column][i] = np.nan
```

```python
# supprime la colonne index et la recrée
def changeIndex(dataFrame, columnName):
    if columnName in dataFrame.columns:
        del dataFrame[columnName]
    if not(columnName in dataFrame.columns):
        dataFrame.insert(0, columnName, 0, allow_duplicates=False)
        for i in range(len(dataFrame)):
            dataFrame[columnName][i] = i
```

```python
# On remplace les valeurs égales à 0 par une chaine de caractères
def notInStock(dataFrame, column):
    for i in range(len(dataFrame)):
        if dataFrame[column][i] == 0:
            dataFrame[column][i] = 'not in stock'
```

```python
# retourne toutes les occurences d'un type d'item donné par l'utilisateur
def getItemType(dataFrame, itemType):
    dataFrame = dataFrame.loc[dataFrame['Item_Type']==itemType]
    return dataFrame
getItemType(df, 'Canned')
```

```python
# Créer 3 catégorie pour le poids
def subCategory(dataFrame, column, value1, value2, value3, palier1, palier2):
        for i in range (len(dataFrame)):
            if type(dataFrame.loc[i, column]) != str :
                if dataFrame.loc[i, column] < palier1 :
                    dataFrame.loc[i, column] = value1
                elif dataFrame.loc[i, column] > palier1 and dataFrame.loc[i, column] < palier2 :
                    dataFrame.loc[i, column] = value2
                else :
                    dataFrame.loc[i, column] = value3
        return dataFrame
                
subCategory(df, 'Item_Weight', '1-10 kg', '11-15 kg', '15 + kg', 10, 15)
```
